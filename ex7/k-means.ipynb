```python
%matplotlib inline

import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import scipy.optimize as op
import math as math
from sklearn import svm, datasets

# use seaborn plotting defaults
import seaborn as sns; sns.set()
```

```python
# Load and visulaize data
data = sio.loadmat('ex7data2.mat')
X = data['X']
plt.scatter(X[:, 0], X[:, 1])
```

```python
numOfCentroids = 3;  
initial_centroids = np.array([[3, 3], [6, 2], [8, 5]])

def findClosestCentroids(X, initial_centroids):
    
    distances = np.zeros([X.shape[0], initial_centroids.shape[0]])
    for centroid_index, centroid in enumerate(initial_centroids):
        column = ((X - centroid)**2).sum(axis=1)
        distances[:, centroid_index] = column
         
    # Find the index of the lowest value on each row. This value is the index of the centroid that is closest
    closestCentroid = np.argmin(distances, axis=1)
        
    return closestCentroid
    
# Print the results for the first 3 examples. Note: The result that the result Andrew Ng gives [1 3 2] is effectivly
# the same as this. The diffence coming from that he is using 1 index arrays and python uses 0 indexing.
closestCentroid = findClosestCentroids(X[:3], initial_centroids)
print("Closest Centroids for the first 3 examples: ", closestCentroid)
```

```python

def computeCentroids(X, closestCentroids, centroids):
    xm, xn = X.shape
    cm, cn = centroids.shape
    
    centroidMeans = np.zeros([cm, xn]) 
    examplesPerCentroid = np.zeros([cm,])
        
    for x_index, x in enumerate(closestCentroids):
        
        for centroid_index, centroid in enumerate(centroids):
            if x == centroid_index:
                centroidMeans[centroid_index] += X[x_index]
                examplesPerCentroid[centroid_index] += 1
                
    return centroidMeans / examplesPerCentroid.reshape([cm, 1])

# Initialize the centroids to be random examples
def kMeansInitCentroids(X, numberOfCentroids):
    
    # Randomly reorder the indices of examples
    shuffled_X = np.random.permutation(X)
    
    # Take the first K examples as centroids
    return(shuffled_X[:numberOfCentroids])

# Performs a single itterating of the kmeans algorithm
def kmeans(X, centroids):
    
    closestCentroids = findClosestCentroids(X, centroids)
    updated_centroids = computeCentroids(X, closestCentroids, centroids)
    
    return updated_centroids     
```

```python
def startKmeans(X, numOfCentroids, ittr, visualize):
    
    #initial_centroids = np.array([[3, 3], [6, 2], [8, 5]])
    initial_centroids = kMeansInitCentroids(X, numOfCentroids)
    print(initial_centroids.shape)


    if visualize:
        # Visualize all the data and the starting positions of the cluster centers
        plt.scatter(X[:, 0], X[:, 1])
        plt.scatter(initial_centroids[:,0], initial_centroids[:, 1], color='red')
        plt.title("Kmeans algorithm after itteration: %d" %ittr)
        
    
    for i in range(ittr):

        # Where the actual algorithm is
        updated_centroids = kmeans(X, initial_centroids)
        
        if visualize:
            lines = np.vstack(zip(initial_centroids, updated_centroids))
            for line_index in range(0, len(lines), 2):
                line = np.vstack((lines[line_index], lines[line_index+1]))
                plt.plot(line[:,0], line[:, 1], 'k-', lw=1)

            # add the new updated cluster centers to the plot
            plt.scatter(updated_centroids[:,0], updated_centroids[:, 1], color='red')
            
        initial_centroids = np.copy(updated_centroids)
        
    # At this point the initial_centroids variable has been updated by itter number of kmeans itterations
    return initial_centroids
    
```

```python
improved_centroids = startKmeans(X,3, 7, True)
```

```python
# flatten the pixels down so the algorithm can use it
bird = bird.reshape(16384, 3)
colours = startKmeans(bird, 16, 30, False)

# Find which pixels are closest to which colour cluster 
newAssignments = findClosestCentroids(bird, colours)

# Change each pixel to the colour of the cluster center it belongs to. Taking the number of unique colours in the 
# image down to 16
for pixel_index, newColour in enumerate(newAssignments):
    bird[pixel_index] = colours[newColour]

    
# unflatten the image back to it's original shape so that it can be displayed    
bird = bird.reshape([128, 128, 3])

# Display the compressed 16 colour image
plt.imshow(bird)
```
