```python
%matplotlib inline

import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import scipy.optimize as op
import math as math
from sklearn import svm, datasets

# use seaborn plotting defaults
import seaborn as sns; sns.set()
```

```python
# Load and visulaize data
data = sio.loadmat('ex7data1.mat')
X = data['X']
plt.scatter(X[:, 0], X[:, 1])
```

```python
# Used to standardize the data. Returns examples with every feature having a mean of 0 and a 
# standard deviation of 1. The 'Train' version version of this function is only used on the training examples.
def featureNormalizeTrain(X):
    featureMeans = np.array([np.mean(x, dtype=np.float64) for x in X.T])
    scaledX = X - featureMeans
    standardDeviations = np.array([np.std(x, dtype=np.float64) for x in scaledX.T])
    standardizedX = scaledX / standardDeviations

    return standardizedX, featureMeans, standardDeviations

X, featureMeans, standardDeviations = featureNormalizeTrain(X)
# Plot all the data
plt.scatter(X[:, 0], X[:, 1])

def PCA(X):
    sigma  = np.dot(X.T, X) / X.shape[0]
    U, S, V = np.linalg.svd(sigma)
    return U, S, V
    
U, S, V = PCA(X)

print(U.shape)

print(U[0, 0])
print(U[1, 0])
# This should be [0, 0] because the data has been normalized and scaled
featureMeans = X.mean(axis=0)

plottingPoint = featureMeans + 1.5 * S[0] * U[:, 0].T
x_line = np.array([featureMeans[0], plottingPoint[0]])
y_line = np.array([featureMeans[1], plottingPoint[1]])
plt.plot(x_line, y_line, 'k-', lw=1)

plottingPoint = featureMeans + 1.5 * S[1] * U[:, 1].T
x_line = np.array([featureMeans[0], plottingPoint[0]])
y_line = np.array([featureMeans[1], plottingPoint[1]])
plt.plot(x_line, y_line, 'k-', lw=1 )



```

```python
def projectData(X, U, K):
    Ureduce = U[:, :K]
    X_red = np.dot(Ureduce.T, X.T)
    return X_red.T
   
def recoverData(Z, U, K):
    Ureduce = U[:, :K] 
    X_rec = np.dot(Ureduce, Z.T)
    return X_rec.T

# Project the data set X on 1 dimension
z = projectData(X, U, 1)
print("Projection of the first example: {}".format(z[0]))

# Recover an approximation to the dataset X from z
X_rec = recoverData(z, U, 1)  
print("Approximation of the first example:{}, {}".format(X_rec[0, 0], X_rec[0, 1]))

# Visualize what pca is doing to each point in dataset X
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(X_rec[:, 0], X_rec[:, 1], color='red')
lines = np.vstack(zip(X, X_rec))
for line_index in range(0, len(lines), 2):
    line = np.vstack((lines[line_index], lines[line_index+1]))
    plt.plot(line[:,0], line[:, 1], 'k-', lw=1)

    
        
```

```python
# Load and visulaize data
data = sio.loadmat('ex7faces.mat')
X = data['X']

# Needs values inverting to be correctly displayed
X *= -1

X_norm, featureMeans, standardDeviations = featureNormalizeTrain(X)



def displayData(X, numOfSubplots):
    
    numOfImages = X.shape[0]
        
    # set up the figure
    fig = plt.figure(figsize=(6, 6))  # figure size in inches
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
    
    
    
    #for i, image in enumerate(np.random.randint(numOfImages, size=numOfSubplots)):
    for i in range(numOfSubplots):

        # Convert raw vector to a 32x32 array so that it can be printed. 
        imageData = X[i]
        printableImageArray = np.reshape(imageData, (-1, 32)).T

        # Add image of digit to the plot
        ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
        ax.imshow(printableImageArray, cmap=plt.cm.binary)

        


displayData(X, 64)
        
    
```

```python
# Run PCA on face data
U, S, V = PCA(X_norm)

#X_red = projectData(X, U, )

principle_componants = U[:, :64].T
print(principle_componants.shape)
displayData(principle_componants, 64)
```

```python
X_reduced = projectData(X_norm, U, 100)
X_recovered = recoverData(X_reduced, U, 100)

print("Original and Recovered Images")
displayData(X, 16)
displayData(X_recovered, 16)
```
