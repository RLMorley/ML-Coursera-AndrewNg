```python
%matplotlib inline

import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import scipy.optimize as op
import math as math



# use seaborn plotting defaults
import seaborn as sns; sns.set()
```

```python
data = sio.loadmat('ex3data1.mat')
X = data['X']
y = data['y']
y = y.flatten()

```

```python
def displayData():
        
    # set up the figure
    fig = plt.figure(figsize=(6, 6))  # figure size in inches
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
    
    for i, image in enumerate(np.random.randint(5000, size=64)):
        
        # Convert raw vector to a 20x20 array so that it can be printed
        imageData = X[image]
        printableImageArray = np.reshape(imageData, (-1, 20)).T

        # Add image of digit to the plot
        ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
        ax.imshow(printableImageArray, cmap=plt.cm.binary)
        
        # label the image with the target value
        ax.text(0, 17, str(y[image]))
        
displayData()
```

$$ h_\theta(x^{(i)}) = g(\theta^Tx^{(i)}) $$ and

$$g(z) = \frac{1}{1+e^{-z}} $$

```python
# The raw result of the logistic regression. Returns a real value between 0 and 1
def hypothesis(theta, X):
    # Sigmoid function
    #print("hypo theta: ", theta.shape)
    #print("hypo X: ", X.shape)

    return (1 / (1 + math.e**( -1 * ( np.dot(X, theta) ) )))
```

Regularized Cost Function for Logistical Regression: $$ J(\theta) = \frac{1}{m}\sum_{i=1}^m [-y^{(i)}\log{(1-h_\theta(x^{(i)})}) - (1-y^{(i)})\log{(1-h_\theta(x^{(i)})} ] + \frac{\lambda}{2m} \sum_{j=1}^n\theta_j^2 $$

```python


# Regularized cost function for logistic regression
def costFunction(theta, X, y, lamda):
    #hypoth = hypothesis(theta, X)
    #m = X.shape[0] * 1.0
    
    m = len(y) * 1.0
    hypoth = hypothesis(theta, X)
           
    rawCost =(1/m) * ((-y * np.log(hypoth)) - ((-y + 1) * np.log(-hypoth + 1))).sum() 
    regularizationTerm = sum((theta ** 2)[1:]) * float(lamda) / (2 * m) 
    
    cost = rawCost + regularizationTerm
    return cost

# Regularized gradient function for logistic regression
def gradient(theta, X, y, lamda):
        
    gradientDerivitive = (((hypothesis(theta, X)) - y).dot(X)) * (1 / X.shape[0])
    regularizationDerivitive = (lamda / X.shape[0]) * theta
        
    gradient = gradientDerivitive.T + regularizationDerivitive
    gradient[0] -= regularizationDerivitive[0]

    # Flattern ensures that the return value is a 1-d numpy array. aka a vector 
    return gradient.flatten()

# Returns the classification prediction for the learnt model. Returns binary value of 0 or 1
def predict(X):
    sigmoidPrediction = hypothesis(optimal_theta.T, X) 
    rounded = np.rint(sigmoidPrediction)
    
    # Rounding the array should solve most of the classification, but a few will not. For those
    # the largest value will be searched for and set as the class
    for index, result in enumerate(rounded):
        if result.sum() != 1:
            ind = sigmoidPrediction[index].argmax()
            rounded[index][ind] = 1
            
    return rounded
        
        







```

```python
def trainMultiClassRegression(numOfClasses, X, y):
    theta = np.zeros([numOfClasses, X.shape[1]])
    for currentClass in range(numOfClasses):
        
        singleClassY = np.zeros([y.shape[0],])
        
        for index, label in enumerate(y):
            if label == currentClass:
                singleClassY[index] = 1
            else:
                singleClassY[index] = 0
                
        theta[currentClass] = trainSingleClassRegression(X, singleClassY, 1)
        
    return theta
                
               
def trainSingleClassRegression(X, y, lamda):
    m, n = X.shape
    initial_theta = np.zeros(n)
    lamda = 1.0 # Regularization Parameter

    myargs = (X, y, lamda)
    opts = {'full_output': True, 'maxiter': 600}

    #optimal_theta, cost, grad_at_min, inv_hessian_matrix,\
    #fun_calls, grad_calls, warn_flags = op.optimize.fmin_bfgs(costFunction,
    #                                    initial_theta,
    #                                    args=myargs,
    #                                    fprime=gradient,
    #                                    **opts)
    
  

    # fmin_cg is like fmin_bfgs but is much faster on high dimensional data
    optimal_theta = op.optimize.fmin_cg(f = costFunction, x0 = initial_theta, args = myargs, fprime = gradient, maxiter = 600)
    
    return optimal_theta
   
        
        
    #return optimal_theta

# Add the column of 1's to the training data
ones = np.ones([X.shape[0], 1])
Xone = np.hstack((ones, X))

# Data labels in y use '10' for 0, '1' for 1, '2' for 2 and so on. This changes '10' for 0 to '0' for better code 
for i, label in enumerate(y):
    if label == 10:
        y[i] = 0

optimal_theta = trainMultiClassRegression(10, Xone, y)




```

```python
# Gets predictions of learnt model on entire training data set and compares to label vector (y) for success rate
def trainingSetPerformance():
    results = predict(Xone)
    
    correctPredictions = 0
    for index, label in enumerate(y): 
        if results[index, label] == 1:
            correctPredictions += 1
        #else:
            #print('Incorrect prediction at index: ', index)
            
    print("Number of Incorrect Predictions: ", y.size - correctPredictions)
    return correctPredictions / y.size
            
        
    
    #differences = np.absolute((y - results))
    #return 1 - (differences.sum() / y.size)
    
print("Model Success Rate on Training Data:", trainingSetPerformance()*100, "%")
    
```
