```python
data = sio.loadmat('ex4data1.mat')
rawX = data['X']
y = data['y']
y = y.flatten()

# Add the column of 1's to the training data
ones = np.ones([rawX.shape[0], 1])
X = np.hstack((ones, rawX))

# Data labels in y use '10' for 0, '1' for 1, '2' for 2 and so on. This changes '10' for 0 to '0' for better code 
for i, label in enumerate(y):
    if label == 10:
        y[i] = 0
       
convertedY = np.zeros([X.shape[0], 10])    
for i, label in enumerate(y):
    convertedY[i, label] = 1
    
# NN details without bias units
inputLayerSize = 400
hiddenLayerSize = 25
outputLayerSize = 10


# Load provided theta values
theta = sio.loadmat('ex4weights.mat')
importTheta1 = theta['Theta1']
importTheta2 = theta['Theta2']

print(importTheta1.shape)
print(importTheta2.shape)
```

we define the error $ \delta_j^l $ of neuron j in layer l by $ \delta_j^l = \frac{\partial C}{\partial z_j^l} $

For each node in the output layer set $$\delta = (a_k^{(3)} - y_k) $$

The cost function we are using is: $$ J(\theta) = \frac{1}{m}\sum_{i=1}^m [-y^{(i)}\log{(1-h_\theta(x^{(i)})}) - (1-y^{(i)})\log{(1-h_\theta(x^{(i)})} ] + \frac{\lambda}{2m} \sum_{j=1}^n\theta_j^2 $$


```python
# The raw result of the logistic regression. Returns a real value between 0 and 1
def hypothesis(theta, X):
    # Sigmoid function
    #print("hypo theta: ", theta.shape)
    #print("hypo X: ", X.shape)
    return (1 / (1 + math.e**( -1 * ( np.dot(X, theta) ) )))

def pack_thetas(theta1, theta2):
        return np.concatenate((theta1.flatten(), theta2.flatten()))
    
def unpack_thetas(thetas):
    t1_start = 0
    t1_end = hiddenLayerSize * (inputLayerSize + 1)
    theta1 = thetas[t1_start:t1_end].reshape((hiddenLayerSize, inputLayerSize + 1))
    theta2 = thetas[t1_end:].reshape((outputLayerSize, hiddenLayerSize + 1))
    return theta1, theta2
    

def forwardProp(theta1, theta2, X):
    #print('forwardProp called')
        
    f2 = np.dot(X, theta1.T)

    a2 = sigmoid(f2)
    
    # Add the column of 1's to the hidden layer activation
    ones = np.ones([a2.shape[0], 1])
    a2 = np.hstack((ones, a2))
    
    # Calcs the output layer
    f3 = np.dot(a2, theta2.T)
    a3 = sigmoid(f3)
     
    #output = np.roll(a3, 1, 1)

    return a3, f3, a2, f2

def nnCostFunction(nn_theta, X, y, lamda):

    theta1, theta2 = unpack_thetas(nn_theta)
          
    m = len(y) * 1.0
    a3, f3, a2, f2 = forwardProp(theta1, theta2, X)
   
    currentCostUnRegularized = 0
    for exampleIndex, example in enumerate(X):
            currentCostUnRegularized += ((-convertedY[exampleIndex] * np.log(a3[exampleIndex])) - ((-convertedY[exampleIndex] + 1) * np.log(-a3[exampleIndex] + 1))).sum() 
    
    mainCostTerm = currentCostUnRegularized / X.shape[0]

    regularizationTerm = ((theta1[:, 1:]**2).sum() + (theta2[:, 1:]**2).sum()) * (lamda / (2 * X.shape[0]))
    
    cost = mainCostTerm + regularizationTerm
    return cost

def sigmoid(z):
    return 1/ (1+ math.e**-z)
def sigmoidGradient(z):
    return sigmoid(z) * (1 - sigmoid(z))

def randInitializeWeights(L_in, L_out):
    # Randomly initialize the weights to small values
    epsilon_init = 0.12
    W = np.random.rand(L_out, 1 + L_in) * 2 * epsilon_init - epsilon_init
    
    return W


def backPropVector(nn_theta, X, y, lamda):
    
    # Reshape nn_params into theta 1 and theta 2
    theta1, theta2 = unpack_thetas(nn_theta)
    
    outputActivations, z3, a2, z2 = forwardProp(theta1, theta2, X)
    
    # Delta3
    outputErrors = outputActivations - y
    
    # Delta2 
    hiddenLayerErrors = (outputErrors.dot(theta2))[:, 1:] * sigmoidGradient(z2)
    
    deltaGrad1 = X.T.dot(hiddenLayerErrors)
    deltaGrad2 = a2.T.dot(outputErrors)
    deltaGrad1 = deltaGrad1.T
    deltaGrad2 = deltaGrad2.T
   
    # Divide by numer of examples m. m = X.shape[0]
    deltaGrad1 = (deltaGrad1 / X.shape[0])
    deltaGrad2 = (deltaGrad2 / X.shape[0])
    
    # Add the regularization term    
    deltaGrad1 += theta1 * (lamda/X.shape[0])
    deltaGrad2 += theta2 * (lamda/X.shape[0])

    
    gradients = pack_thetas(deltaGrad1, deltaGrad2)   
    return gradients
    

    
def backProp(nn_theta, X, y, lamda):
    
    # Reshape nn_params into theta 1 and theta 2
    theta1, theta2 = unpack_thetas(nn_theta)
    
    deltaGrad1 = 0
    deltaGrad2 = 0      
    for exampleIndex, example in enumerate(X):
        outputActivations, z3, a2, z2 = forwardProp(theta1, theta2, np.reshape(example, [1,example.shape[0]]))
       
        outputActivations = outputActivations.flatten() # Convert 2d [1, 10] into a 1d vector for ease
   
        outputErrors = np.zeros([outputActivations.shape[0]],)
  
        # Delta3
        for outputUnit, unitActivation in enumerate(outputActivations):
            outputErrors[outputUnit] = unitActivation - convertedY[exampleIndex, outputUnit]
            
        # Delta2       
        hiddenLayerError = ((theta2.T)[1:].dot(outputErrors) * sigmoidGradient(z2).flatten())

        # Skip the first column of 1's
        #deltaGrad1 += np.outer(hiddenLayerError[1:], example) 
        deltaGrad1 += np.outer(hiddenLayerError, example) 
        deltaGrad2 += np.outer(outputErrors, a2)

   
    # Divide by numer of examples m. m = X.shape[0]
    deltaGrad1 = deltaGrad1 / X.shape[0]
    deltaGrad2 = deltaGrad2 / X.shape[0]
    
    
    # Add the regularization term
    #deltaGrad1 += theta1[1:] * (lamda/X.shape[0])
    deltaGrad1 += theta1 * (lamda/X.shape[0])
    deltaGrad2 += theta2 * (lamda/X.shape[0])
    
    # Unravel gradients
    #theta1_grad = deltaGrad1.T.ravel()
    #theta2_grad = deltaGrad2.T.ravel()

    #gradients = np.array([theta1_grad, theta2_grad])
    gradients = pack_thetas(deltaGrad1, deltaGrad2)
         
    return gradients


    

initial_theta1 = randInitializeWeights(400, 25)
initial_theta2 = randInitializeWeights(25, 10)
#print("initial theta1 shape: ", initial_theta1.shape)
#print("initial theta2 shape: ", initial_theta2.shape)

#initial_nn_params = np.array([(initial_theta1.T).flatten(), (initial_theta2.T).flatten()])
initial_nn_params = pack_thetas(initial_theta1, initial_theta2)

lamda = 0.0 # Regularization Parameter
myargs = (X, convertedY, lamda)
#optimal_theta = op.optimize.fmin_cg(f = nnCostFunction, x0 = initial_nn_params, args = (myargs), fprime = backPropVector, maxiter = 200)


#test = backPropVector(initial_nn_params, X, convertedY, 1)
#correct = backProp(initial_nn_params, X, convertedY, 1)





 
```
