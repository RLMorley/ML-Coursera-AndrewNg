```python
%matplotlib inline

import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import scipy.optimize as op
import math as math

# use seaborn plotting defaults
import seaborn as sns; sns.set()
```

```python
data = sio.loadmat('ex5data1.mat')
X_train = data['X']
y_train = data['y'].flatten()

X_test = data['Xtest']
y_test = data['ytest'].flatten()

X_validation = data['Xval']
y_validation = data['yval'].flatten()

# VERY useful function for data preprocessing. Adds a column of 1's at the start of the data.
def addBiasValues(X):
    ones = np.ones([X.shape[0], 1])
    return np.hstack((ones, X))

```

```python
plt.scatter(X_train, y_train)
plt.xlabel('Change in water level (x)')
plt.ylabel('Water flowing out of the damn (y)')
```

```python
class LinearRegression(object):
        
          
    def hypothesis(self, theta, X):
        return np.dot(X, theta).flatten()

    # X data must come with the column of 1's added beforehand
    def linearRegCost(self, theta, X, y, lamda):
        m, n = X.shape
    
        rawCostTerm = ((self.hypothesis(theta, X) - y) ** 2).sum() * (1 / (2 * m))
        
        # Every theta value is regularized EXCEPT theta[0]. i.e the bias weight
        regularizationTerm = (theta[1:] ** 2).sum() * (lamda / (2 * m))
    
        return rawCostTerm + regularizationTerm

    def linearRegGradient(self, theta, X, y, lamda):
        m, n = X.shape
    
        unregularizedGradient = ((self.hypothesis(theta, X) - y).dot(X)) * (1 /  m)  

        regularizationTerm = theta * (lamda / m)
        
        # Do not regularize the bias weight
        regularizationTerm[0] = 0
    
        # Every theta value is regularized EXCEPT theta[0]. i.e the bias weight
        grad = unregularizedGradient + regularizationTerm
        
        return grad
    
    # Expects data in the X variable to be without the column of 1's (bias values)
    def trainLinearRegression(self, X, y, lamda):
        
        # Data Preprocessing
        X = addBiasValues(X)
                
        initial_theta = np.ones([X.shape[1],])
        myargs = (X, y, lamda)
        #self.optimal_theta = op.optimize.fmin_cg(f = self.linearRegCost, x0 = initial_theta \
                                                # , args = myargs, fprime = self.linearRegGradient, maxiter = 3000)
            
        result = op.minimize(self.linearRegCost, initial_theta, args=myargs, jac=self.linearRegGradient, method='BFGS')

        self.optimal_theta = result.x
        
    def predict(self, X):
        return self.hypothesis(self.optimal_theta, X)
        
```

```python
# Create instance of the Linear Regression model
lin = LinearRegression()

# Add column of ones
X = addBiasValues(X_train)

initial_theta = np.ones([X.shape[1],])

print(lin.linearRegCost(initial_theta, X, y_train, 1))
print(lin.linearRegGradient(initial_theta, X, y_train, 1))
```

```python
lin.trainLinearRegression(X_train, y_train, 0)

plt.scatter(X_train, y_train)
plt.xlabel('Change in water level (x)')
plt.ylabel('Water flowing out of the damn (y)')

# Build an X array for plotting the found function. Should be a column of 1's and a range of 0-50
foundFunctionX = np.arange(-50, 50).reshape(100,1)
foundFunctionX = addBiasValues(foundFunctionX)

# Use built X array with the found weights to generate the accociated y values. Then plot on a graph
foundFunctionY = lin.predict(foundFunctionX)
plt.plot(foundFunctionX[:,1], foundFunctionY, "g")
```

```python
# Repeatedly trains the linear regression model with increasingly more data in the training examples set. 
# The training and validation error on each model is calculated and is then ploted on a graph showing the 
# error against number of training examples used in training. The training error is only found using the examples
# actually used in model training. i.e Not the whole X_train. The validation error is however found using the entire
# validation set for every model trained.
def plotLearningCurve(X_train, y_train, X_validation, y_validation, lamda):
    # Arrays to hold the error from each run on each data portion
    error_train = np.array([])
    error_validation = np.array([])
    for index, example in enumerate(X):
    
        # Train the linear regression model 
        X_slice = X_train[:index + 1, :]
        y_slice =  y_train[:index + 1]
            
        lin.trainLinearRegression(X_slice, y_slice, lamda)
    
        # Add the column of ones to the data for the bias term
        X_slice_ones = addBiasValues(X_slice)
    
        # Add the column of ones to the data for the bias term        
        X_validation_ones = addBiasValues(X_validation)
        
        # Add the errors for the current run to their respective arrays for storage    
        error_train = np.append(error_train, lin.linearRegCost(lin.optimal_theta, X_slice_ones, y_slice, 0) )
        error_validation = np.append(error_validation, lin.linearRegCost(lin.optimal_theta, X_validation_ones, y_validation, 0))
     
    # Draw the error graph
    x = np.arange(X.shape[0])
    plt.title('Regularization Parameter (lamda) =  %d' % lamda)
    plt.xlabel("Number of Examples used in Training")
    plt.ylabel("Error")
    plt.plot(x+1, error_train, label="training set")
    plt.plot(x+1, error_validation, label="validation set")
    plt.legend(loc='upper right')    

plotLearningCurve(X_train, y_train, X_validation, y_validation, 0)
    
```

```python
def mapFeatures(X, degree):
    m, n = X.shape 
    
    new_features = np.array([])
    for degree in range(degree):
        degree = degree + 1
        if degree == 1: continue
        new_feature = X[:, 0] ** degree 
        
        X = np.column_stack((X, new_feature))
        
    return X

# Used to standardize the data. Returns examples with every feature having a mean of 0 and a 
# standard deviation of 1. The 'Train' version version of this function is only used on the training examples.
def featureNormalizeTrain(X):
    featureMeans = np.array([np.mean(x, dtype=np.float64) for x in X.T])
    scaledX = X - featureMeans
    standardDeviations = np.array([np.std(x, dtype=np.float64) for x in scaledX.T])
    standardizedX = scaledX / standardDeviations

    return standardizedX, featureMeans, standardDeviations

# This version of the normalize function is to be only used on examples that the ALREADY trained model has not seen.
# As such, it requires the standarization parameters returned by the 'Train' version of this function when it was
# used on the train examples used to train the model.
def featureNormalize(X, featureMeans, standardDeviations): 
    return (X - featureMeans) / standardDeviations
        
```

```python
# This function trains several models each with a different value of lamda (regualization paramater) and then 
# calculates the training and validation for each and returns the value of lamda that gives the lowest validation
# error. It also returns the training and validation errors for the models in 2 arrays. 
def validationCurve(X_train, y_train, X_validation, y_validation):
    lamdaValues = np.array([0, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10])
    lin = LinearRegression()
    
    # Validation set preprocessing    
    X_validation = addBiasValues(X_validation)
    
    
    
    # Validation set preprocessing    
    X_train_ones = addBiasValues(X_train)
    
    crossvalErrors = np.array([])
    trainErrors = np.array([])
    for lamda in lamdaValues:
        
        lin.trainLinearRegression(X_train, y_train, lamda)
               
        crossvalErrors = np.append(crossvalErrors, lin.linearRegCost(lin.optimal_theta, X_validation, y_validation, 0))
        trainErrors = np.append(trainErrors, lin.linearRegCost(lin.optimal_theta, X_train_ones, y_train, 0))

        
    bestLamda = lamdaValues[np.argmin(crossvalErrors)]
    
    return bestLamda, trainErrors, crossvalErrors
        
    
```

```python
# Find the best value of lamda 
bestLamda, trainErrors, crossvalErrors = validationCurve(normalizedX_train, y_train, X_validation_poly, y_validation)
print("Best lamda: ", bestLamda)

# The y values explode near when X is near 0. This is required to force the interesting part of the the plots into
# focus
plt.ylim(0, 20)


# Plot the training and validation errors against the value of lamda used in training
plt.plot(np.array([0, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10]), trainErrors, label="training set")
plt.plot(np.array([0, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10]), crossvalErrors, label="validation set")
plt.xlabel("Lamda")
plt.ylabel("Error")
plt.legend(loc='upper right')   
```
