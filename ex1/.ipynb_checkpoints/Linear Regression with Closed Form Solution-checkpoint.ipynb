```python
import matplotlib.pyplot as plt
%matplotlib inline

# use seaborn for plot defaults
# this can be safely commented out
import seaborn; seaborn.set()
```

```python
# This is the target function
def f(x):
    y = (2*x)-2
    return y

from random import randint
import numpy as np

numberOfPoints = 100

X = np.zeros([numberOfPoints, 2], dtype=np.float_)  # numberOfPoints by 2 matrix
y = np.zeros([numberOfPoints, 1], dtype=np.float_)  # numberOfPoints by 1 matrix/vector

index = 0
while index < numberOfPoints:
    newXa = randint(0,50)
    
    X[index][0] = 1; # Every examples first feature must be 1.
    X[index][1] = newXa;
    

    # Generate the y coordiate by using the target function and a random deviation to simulate noise
    deviation = randint(-20, 20)
    y[index] = f(newXa) + deviation
    
    index += 1
    
targetFunctionX = np.array(range(50))
targetFunctionY = f(targetFunctionX)

import pylab as pl
pl.scatter(X[:, 1], y[:,0])
#pl.scatter(X[:, 0], X[:, 1], c = y, cmap=plt.cm.get_cmap('jet', 2))
#pl.scatter(X[:, 0], X[:, 1], c = y, cmap=plt.cm.get_cmap('nipy_spectral', 10))

pl.plot(targetFunctionX, targetFunctionY, "r")
```

```python
# Weights array initialised to zeros
weights =  np.zeros([2, 1])

# This uses mean squared errors as the error metric. This is often called J(theta) in formal notation
def costFunction(X, y, weights):
    currentCost = 0
    for pointIndex, point in enumerate(X):
        
        # Adding the squared error to the current sum total
        currentCost += (np.dot(point, weights) - y[pointIndex])**2
        
    meanSquaredErrors = currentCost / X.size
    
    return meanSquaredErrors

from numpy.linalg import inv
def solveForTheta(X, y):
    inverse = inv(np.dot(X.transpose(), X))
    theta = np.dot(inverse, np.dot(X.transpose(), y))
    return theta                 

theta = solveForTheta(X, y)
print(theta)

#Draw the line that represents the weight vector.
foundFunctionX = np.arange(50).reshape(50,1)

ones = np.ones([50,1])
theX = np.hstack((ones, foundFunctionX)) # Insert a column of ones
foundFunctionY = np.dot(theX, theta) # Calculate the y values using the found weights/theta

pl.scatter(X[:, 1], y[:,0] )
pl.plot(range(50), foundFunctionY, "g")
    
    
    
    
```
